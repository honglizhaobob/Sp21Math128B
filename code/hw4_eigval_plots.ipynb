```python
import numpy as np
import numpy.linalg as la
import sympy as sym
import scipy as sp

from sympy.matrices import Matrix
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#part a
A = Matrix([
    [4, 0, 1, 3],
    [0, 4, 2, 1],
    [1, 2, -2, 0],
    [3, 1, 0, 4]
])
print(A.is_diagonalizable())
A = np.array(A).astype('float')
W, V = la.eig(A)
```

```python
# display eigenvalues of A
W
```

```python
# plotting
fig, ax = plt.subplots(figsize=(10,10))

ax.set(xlim=(-10, 10), ylim = (-10, 10))
ax.set_aspect(1)
circ1 = plt.Circle((4, 0), 4, color="red", fill=False)
circ2 = plt.Circle((4, 0), 3, color='blue', fill=False)
circ3 = plt.Circle((-2, 0), 3, color='purple', fill=False)
circ4 = plt.Circle((4, 0), 4, color='black', fill=False)
ax.add_artist(circ1); ax.add_artist(circ3)
ax.add_artist(circ2);
ax.axhline(0, label='Re',color='black');
tick_size = 100
plt.scatter(W[0], 0, marker='x', s=tick_size, color='red');
plt.scatter(W[1], 0, marker='x', s=tick_size, color='blue');
plt.scatter(W[2], 0, marker='x', s=tick_size, color='purple'); 
plt.scatter(W[3], 0, marker='x', s=tick_size, color='black');
plt.grid();
ax.axvline(1, label='x=1'); plt.legend();
abs(W), max(abs(W))
```

```python
#partb
A = Matrix([
    [1, 0, -1, 1],
    [2, 2, -1, 1],
    [0, 1, 3, -2],
    [1, 0, 1, 4]
])
A = np.array(A).astype('float')
W, V = la.eig(A)
```

```python
W
```

```python
# plotting
fig, ax = plt.subplots(figsize=(10,10))

ax.set(xlim=(-10, 10), ylim = (-10, 10))
ax.set_aspect(1)
circ1 = plt.Circle((1, 0), 2, color="black", fill=False)
circ2 = plt.Circle((2, 0), 4, color="purple", fill=False)
circ3 = plt.Circle((3, 0), 3, color='orange', fill=False)
circ4 = plt.Circle((4, 0), 2, color='green', fill=False)
ax.add_artist(circ1); ax.add_artist(circ3);
ax.add_artist(circ2); ax.add_artist(circ4);
ax.axhline(0, label='Re',color='black');
tick_size = 100
plt.scatter(W[0].real, W[0].imag, marker='x', s=tick_size, color='black');
plt.scatter(W[1].real, W[1].imag, marker='x', s=tick_size, color='purple');
plt.scatter(W[2].real, W[2].imag, marker='x', s=tick_size, color='orange'); 
plt.scatter(W[3].real, W[3].imag, marker='x', s=tick_size, color='green');
plt.grid();
ax.axvline(2, label='x=2'); plt.legend();

abs(W), max(abs(W))
```
